## Exploratory Analysis and Visualization

Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
 # Setting the initial setup for the seaborn and matplotlib as required for us
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

## Asking and Answering Questions

Lets dig deeper and answer some interesting questions which are important to be known



#### Q1:On an average, who earn more: male or female?

```python
average_salary_by_gender = salary_df.groupby('Gender')['Salary'].mean()
average_salary_by_gender
```

On an average, male earn comparatively more than female with a small margin

#### Q2: Do the specialization matter for getting good salary, and which specialization?

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Specialization', y='Salary', data=salary_df)
plt.xticks(rotation=45, ha='right')
plt.show()
```

Though there are some outliers in the plot, but we can observe that specialization plays a crucial role in determining the salary

#### Q3: Also, does our collegetier and college performance(CollegeGPA) play a crucial role in determing our salary

```python
plt.figure(figsize=(15, 15))
sns.scatterplot(x='collegeGPA', y='Salary', hue='Degree', palette='inferno', data=salary_df)
plt.show()
```

CollegeGPA don't impact a lot given that you have a decent CollegeGPA above 60.

```python
sns.barplot(x = 'CollegeTier', y ='Salary', data = salary_df);
```

Yet, collegetier has a pivotal role in determining the salary of students

```python
average_salary_by_tier = salary_df.groupby('CollegeTier')['Salary'].mean()

print(average_salary_by_tier)
```

#### Q4: What is the top three domain/specialization chosen by high-earning graduates

```python
average_salary_by_specialization = salary_df.groupby('Specialization')['Salary'].mean()

# Sort the specializations by average salary in descending order
sorted_specializations = average_salary_by_specialization.sort_values(ascending=False)

# Extract the top three specializations
top_three_specializations = sorted_specializations.head(3)

# Display the result
print(top_three_specializations)
```

Computer Networking, Information & Communication Technology and Chemical Engineering are the favourites of high-earning graduates.

#### Q5: What is the influence of other skills apart from the college education on the salary?

```python
plt.figure(figsize=(30, 30))
for i, col in enumerate(salary_df.select_dtypes('number').columns[-6:-1]):
    plt.subplot(7, 3, i % 21 + 1)  # Ensure num is within the valid range
    sns.scatterplot(x=col, y='Salary', data=salary_df)
plt.show()


```

```python
personality_traits = ['conscientiousness', 'agreeableness', 'extraversion', 'nueroticism', 'openess_to_experience']
correlation_matrix = salary_df[personality_traits + ['Salary']].corr()

# Extract correlation values for each personality trait with salary
correlation_with_salary = correlation_matrix['Salary'][personality_traits]

# Display the results
print(correlation_with_salary)
```

There is a negative correlations for conscientiousness,extraversion and neuroticism with salary; while agreeablenss and openess_to_experience do not.

Also, from above plot, we can infer that experience and agreeableness matters more.
